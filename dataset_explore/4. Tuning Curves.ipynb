```python
import matplotlib.pyplot as plt
import numpy as np
import pynapple as nap
```

```python
data_file = "000582/sub-10073/sub-10073_ses-17010302_behavior+ecephys.nwb"
```

```python
data = nap.load_file(data_file)
```

```python
data
```

```python
spikes = data['units']
x = data['SpatialSeriesLED1']['x']
y = data['SpatialSeriesLED1']['y']
```

```python
tuning_curves = nap.compute_1d_tuning_curves(spikes, x, 50)
for i in spikes:
    plt.subplot(4, 2, i + 1)
    plt.plot(tuning_curves[i])
plt.tight_layout()
```

```python
tuning_curves = nap.compute_1d_tuning_curves(spikes, y, 50)
for i in spikes:
    plt.subplot(4, 2, i + 1)
    plt.plot(tuning_curves[i])
plt.tight_layout()
```
